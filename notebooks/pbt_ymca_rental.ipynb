# Financial Calculations Analysis

## Overview
This notebook is designed to analyze the financial changes after incorporating tax. The document focuses on understanding the difference in cost for services per hour, per session, and on a monthly basis. It also highlights the percentage increase in the monthly cost due to these changes.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables
current_price_per_hour = 300  # in CZK
new_price_per_hour = 420  # in CZK
tax_rate = 0.21  # 21%
hours_per_week = 3
weeks_per_month = 4  # Assuming 4 weeks in a month

# Calculate the cost per hour including tax
current_price_per_hour_with_tax = current_price_per_hour * (1 + tax_rate)
new_price_per_hour_with_tax = new_price_per_hour * (1 + tax_rate)

# Calculate the cost for one session (2 hours)
current_cost_per_session = current_price_per_hour_with_tax * hours_per_week
new_cost_per_session = new_price_per_hour_with_tax * hours_per_week

# Calculate the monthly cost
current_monthly_cost = current_cost_per_session * weeks_per_month
new_monthly_cost = new_cost_per_session * weeks_per_month

# Create the labels and values for plotting
labels = ['Current Rate', 'New Rate']
monthly_costs = [current_monthly_cost, new_monthly_cost]

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(labels, monthly_costs, color=['blue', 'red'])
plt.xlabel('Rate Type')
plt.ylabel('Monthly Cost (CZK)')
plt.title('Comparison of Monthly Room Rental Costs')
plt.xticks(rotation=0)
plt.yticks(np.arange(0, max(monthly_costs) + 500, 500))

# Annotate the bars
for i, v in enumerate(monthly_costs):
    plt.text(i, v + 20, str(round(v, 2)), ha='center')

# Show the plot
plt.show()

current_monthly_cost, new_monthly_cost
```

```python
# Calculate the percentage increase in the monthly cost
percentage_increase = ((new_monthly_cost - current_monthly_cost) / current_monthly_cost) * 100

# Forecast the costs over 6 to 12 months for both the current and new rates
months_range = np.arange(6, 13)  # 6, 7, 8, ..., 12
current_costs_forecast = current_monthly_cost * months_range
new_costs_forecast = new_monthly_cost * months_range

# Create the plot for the forecast
plt.figure(figsize=(10, 6))
plt.plot(months_range, current_costs_forecast, label='Current Rate', marker='o')
plt.plot(months_range, new_costs_forecast, label='New Rate', marker='x')
plt.xlabel('Months')
plt.ylabel('Total Cost (CZK)')
plt.title('Forecast of PBT Room Rental Costs Over 6-12 Months')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

percentage_increase
```


## 1. Cost Per Hour Including Tax

### Objective:
To determine the cost per hour after adding the tax to the current and new prices.

### Formula:
\[ \text{Price With Tax} = \text{Current Price} \times (1 + \text{Tax Rate}) \]

### Results:
- Current Price: 363 CZK
- New Price: 508.2 CZK

## 2. Cost for One Session (2 Hours)

### Objective:
To calculate the cost for a two-hour session using the per-hour cost (inclusive of tax).

### Formula:
\[ \text{Session Cost} = \text{Price With Tax} \times 2 \]

### Results:
- Current Session Cost: 726 CZK
- New Session Cost: 1016.4 CZK

## 3. Monthly Cost Calculation

### Objective:
To evaluate the total monthly cost, assuming 4 sessions are held in a month.

### Formula:
\[ \text{Monthly Cost} = \text{Session Cost} \times 4 \]

### Results:
- Current Monthly Cost: 2904 CZK
- New Monthly Cost: 4065.6 CZK

## 4. Percentage Increase in Monthly Cost

### Objective:
To determine the percentage increase in the monthly cost due to the new price.

### Formula:
\[ \text{Percentage Increase} = \frac{\text{New Monthly Cost - Old Monthly Cost}}{\text{Old Monthly Cost}} \times 100 \]

### Results:
Percentage Increase: 40%

## 5. Forecast for 6 to 12 Months

### Objective:
To project the cost for each month from 6 to 12 months using either the current or new rate.

### Notes:
The forecast will rely on the rate provided (either current or new) and will multiply the respective monthly cost by the month number in the range of 6 to 12.
